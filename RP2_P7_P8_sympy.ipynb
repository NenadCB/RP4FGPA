### Omejitve simbolov

V praksi je matematični model način predstavitve sistema, ki ga želimo analizirati. Vsebinsko to pomeni, da neznanke ali simboli, ki jih v modelu uporabljamo, predstavljajo določene značilnosti in količine opazovanega ali analiziranega sistema. Iz tega izhaja, da običajno poznamo nekatere značilnosti uporabljenih simbolov in neznank. Na primer, za določen parameter sistema vemo, da ima lahko le pozitivno vrednost ali pa da nas zanimajo samo realne vrednosti parametra kompleksne pa ne. V teh primerih je smiselno že pri definiciji simbolov te omejitve določiti. CAS sistem bo lahko mnogo bolj učinkovito izvajal operacije nad simbolnimi izrazi.

__Primer:__ 

Izraz $x^a\cdot y^a = (x\cdot y)^a$ je veljaven le, če sta $x$ in $y$ pozitivni števili in če je vrednost $a\in \mathbb{R}$. Zato je smiselno te omejitve določiti v naprej. SymPy bo namreč pri računanju to pretvorbo uporabil le, če simboli ustrezajo kriterijem pravilnosti. Če $x$ in $y$ ne bomo definirali kot simbola, ki lahko zavzameta le pozitivne vrednosti, SymPy pretvorbe v $(x\cdot y)^a$ ne bo naredil in bo izraz pustil v obliki $x^a\cdot y^a$.    

##### Primer brez omejitev vrednosti simbolov

Če vrednosti uporabljenih simbolov ne omejimo, privzeto lahko vsak simbol pripada množici $\mathbb{C}$.

````python
x, y, a = sp.symbols('x y a')
z = x**a * y**a
z
````

Ker v splošnem za vsa kompleksna števila ne velja $x^a\cdot y^a = (x\cdot y)^a$, poenostavitev ne moremo izvesti. 

````python
z.simplify()
````

##### Primer z omejitvami vrednosti simbolov

````python
x, y = sp.symbols('x y', positive=True)
a = sp.symbols('a', real=True)
z = x**a * y**a
z
````

Vrednost izraza $z$ je navidezno enaka primeru zgoraj. Vendar je zaradi omejitev vrednosti simbolov, ki zagotavljajo veljavnost poenostavitve, rezultat funkcije `sympy.simplify()` drugačen.

````python
z.simplify()
````

## Poenostavitve izrazov

SymPy omogoča različne poenostavitve algebraičnih izrazov. Najpreprosteje izraz poenotavimo s funkcijo `sympy.simplify()`. Včasih ta poenostavitev ne vrne najboljše možne poenostavitve. Zato je velikokrat bolj učinkovito, če uporabimo bolj specifične poenostavitvene funkcije. Nekatere bomo predstavili v primerih spodaj.

Primeri:
- __simplify__ - poenostavitev izraza
- __factor__ - razcep racionalne funkcije na faktorje 
- __expand__ - razširitev izraza v kanonično obliko
- __cancel__ - preoblikovanje racionalne funkcije v standardno kanonično obliko $\frac{p}{q}$
- __apart__ - razcep racionalne funkcije na vsoto parcialnih ulomkov

Najprej pa si poglejmo dva primera  uporabe `sympy.simplify()`

#### simplify()

````python
z = sp.sin(x)**2 + sp.cos(x)**2
z
````

````python
z.simplify()
````

````python
z = (x**2 + 3*x + 2)/(x+1)
z.simplify()
````

Vendar v nekaterih primerih funkcija ne naredi, kar bi morda pričakovali.

````python
z = x**2 + 3*x +2
z.simplify()
````

Takrat lahko porabimo bolj specifično funkcijo. V navedenem primeru bi poenostavitev lahko pomenila, da polinom razstavimo na faktorje. V našem primeru nam preoblikovanje pokaže realne nične polinoma (-1 in -2).

#### factor()

````python
z = x**2 + 3*x +2
z.factor()
````

Obratna operacija od razcepljanja na faktorje je razširjanje izraza, kar naredimo s funkcijo `sympy.expand()`.

#### expand()

````python
z = (x+1)*(x+2)
z.expand()
````

````python
z = sp.sin(x + 2) * sp.cos(x - 3)
z.expand(trig=True)  # funkciji povemo, da gre za razširjanje trigonometričnih funkcij
````

#### cancel()
Pri preoblikovanju izrazov in izvajanju operacij, kot npr. odvajanje, kot rezultat večkrat dobimo velike in nepregledne ulomke. V teh primerih je navadno uporabno, da poiščemo skupne faktorje v števcu in imenovalcu ter jih pokrajšamo. Prav temu je namenjena funkcija `sympy.cancel()`.

### Odvod implicitno podane funkcije

V matematiki funkcije navadno podamo v eksplicitni obliki, kjer je ena spremenljivka izražena z drugo, npr. $y = f(x)$. Včasih odnosa med spremenljivkama ni smiselno ali je zapleteno preoblikovati v takšno obliko. Takrat lahko povezanost med spremenljivkama izrazimo v implicitni obliki z uporabo splošne enačbe $F(x, y(x))=0$. Vrednosti $x$ in $y$ seveda nista medsebojno neodvisni.

Odvod implicitne funkcije po spremenljivki, ki jo obravnavamo kot neodvisno (v našem primeru $x$), izračunamo kot razmerje parcialnih odvodov $F(x,y)$ po $x$ in po $y$:
$$y\prime = - \frac{\frac{\partial F(x,y)}{\partial x}}{\frac{\partial F(x,y)}{\partial y}}$$

Iz navedenega izhaja, da lahko odvod implicitne funkcije izpeljemo z običajnim odvajanjem po eni in po drugi spremenljivki. Videli smo že, da v SymPy za to uporabimo funkcijo `sympy.diff()`.

SymPy pa ponuja še alternativno možnost, in sicer funkcijo `sympy.idiff()`, ki neposredno izračuna odvod implicitno podane funkcije. Funkcijo je potrebno preoblikovati v izraz $F(x, y(x))=0$.

Poglejmo oba izračuna na primeru funkcije:
$$\sin y + xy^2 = \frac{1}{2}$$

Najprej preoblikujemo funkcijo v obliko $F(x,y)=0$ in izraz zapišemo v spremenljivko.

````python
x, y = sp.symbols('x y')
f = sp.sin(y) + x*y**2 - sp.Rational(1, 2)
f
````

Izračunamo odvod $F$ po $x$.

````python
dFdx = sp.diff(f, x)
dFdx
````

Izračunamo odvod $F$ po $y$.

````python
dFdy = sp.diff(f, y)
dFdy
````

Uporabimo obrazec za implicitno odvajanje.

````python
dydx = - dFdx/dFdy
dydx
````

## Reševanje enačb

V inženirstvu se pogosto srečamo z reševanje enačb – od preprostih linearnih izrazov do nelinearnih enačb in večspremenljivskih sistemov. Knjižnica SymPy nam na tem področju ponuja zmogljiva orodja, ki med drugim omogočajo:

- algebraično računanje in manipulacije z izrazi
- iskanje ničel funkcij,
- reševanje sistemov enačb (linearnih, nelinearnih, diferencialnih),
- delo s parametri in simbolnimi konstantami,
- analizo rešitev (npr. preverjanje, poenostavljanje, substitucije).

Prednost uporabe simbolnega računanja je v tem, da so rezultati točni, brez zaokroževanja in numeričnih napak in ne približni kot pri uporabi numeričnih metod.  

V nadaljevanju so predstavljeni primeri uporabe SymPy pri delu z enačbami.


### Algebraično računanje

SymPy omogoča pretvarjanje algebraičnih izrazov in računanje s simboli. Izraze predstavimo kot enačbe. V SymPy enačbe shranimo v spremenljivke podatkovnega tipa `Eq`. S funkcijo `solve()` lahko simbolno rešujemo enačbe. V naslednjem preprostem primeru bomo pokazali, kako funkcija preoblikuje enačbo in izpostavi iskano neznanko. 

Primer:

Za palico dolžine $L$ in presekom $A$, za katero je podan modul elastičnosti $E$, izračunaj dopustno osno silo $N$, če je podan največji dovoljen raztezek $\Delta L_{max}$.

Raztezek izračunamo iz enačbe
$$\Delta L = \frac{N \cdot L}{E \cdot A}$$


Definiramo vse simbole, ki jih potrebujemo za zapis enačbe.

````python
N, L, E, A, dL_max = sp.symbols('N, L, E, A, dL_max', positive=True)
````

Ustvarimo enačbo, za kar uporabimo podatkovni tip `Eq`, kjer je prvi parameter izraz na levi strani enačbe in drugi izraz na desni strani.

````python
eq = sp.Eq(dL_max, (N*L)/(E*A))
eq
````

Enačbo rešimo za neznanko $N$.

````python
N_max = sp.solve(eq, N)
N_max[0]
````

Iz primera vidimo, da funkcija `solve()` vrne seznam rezultatov, čeprav je v našem primeru rešitev samo ena. Izpisali smo prvi element seznama rešitev.

### Iskanje ničel funkcije

Funkcijo `solve()` lahko uporabimo za iskanje ničel podane funkcije. Iščemo vrednosti $x$, kjer je $f(x)=0$. 

Primer:

Prosto ležeči nosilec dolžine L je preprosto podprt na obeh krajih in je obremenjen z enakomerno obtežbo q. Poišči kje vzdolž nosilca je upogibni moment $M$ enak 0, če se $M$ izračuna kot:
$$M(x) = \frac{qx}{2}(L-x)$$

Rezultat je iz enačbe sicer očiten. Vidimo, da je $M(x)=0$ za $x=0$ in $x=L$. Izračunajmo rezultate še s SymPy.

Enačbo moramo zapisati tako, da kot prvi parameter zapišemo funkcijski predpis, drugi pa je vrednost 0. Nato enačbo rešimo za vrednost neznanke $x$. 

>OPOMBA: 
>Kadar rešujemo enačbo podano v obliki $f(x)=0$, lahko v funkciji `solve()` kot prvi parameter podamo kar funkcijski predpis in ni potrebno ustvariti objekta `Eq` (glej nalogo spodaj).

````python
x, q, L = sp.symbols('x q L', real=True)
M = (L-x)*q*x/2
eq = sp.Eq(M, 0)
koreni = sp.solve(eq, x)
koreni
````

## Reševanje sistema enačb

Analogno reševanju ene enačbe bomo funkcijo `solve()` uporabili tudi za reševanje sistema enačb.

### Sistem linearnih enačb

Primer:

(naloga 8, str. 374) Zapiši enačbo kvadratne funkcije, ki poteka skozi točke $A(1,3)$, $B(2, -1)$ in $C(-1,2)$.

Sistem treh enačb dobimo, če v enačbo kvadratne funkcije $y=ax^2+bx+c$ vstavimo vse tri točke.

$$3=1^2a+1b+c$$
$$-1=2^2a+2b+c$$
$$2=-1^2a-1b+c$$

Koeficiente funkcije izračunamo z rešitvijo sistema enačb.

Zapišimo enačbe z uporabo SymPy. 

````python
a, b, c = sp.symbols('a b c')
e1 = sp.Eq(3, a+b+c)
e2 = sp.Eq(-1, 4*a + 2*b + c)
e3 = sp.Eq(2, a-b+c)
````

Rešimo sistem enačb za $a$, $b$ in $c$. Rešitev je slovar v katerem so naštete vse spremenljivke in njihove vrednosti.

````python
res = sp.solve([e1, e2, e3], [a, b, c])
````

````python
res
````

Uporabimo rešitve za zapis funkcije $f$.

````python
f = res[a]*x**2 + res[b]*x + res[c]
f
````

Grafično preverimo, ali funkcija resnično poteka skozi podane točke.

````python
fig, ax = plt.subplots(figsize=(3,3))
ax.scatter([1, 2, -1], [3, -1, 2])
sp.plot(f,fig=fig, ax=ax);
````
